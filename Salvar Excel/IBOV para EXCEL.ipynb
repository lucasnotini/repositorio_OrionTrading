```python
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

```python
start = '02-08-2022'
end = dt.datetime.now()
stocks = '^bvsp'
```

```python
df = web.get_data_yahoo(stocks, start)
#remove colunas a não ser fechamento
df = df.drop(['High','Low', 'Open','Volume','Adj Close'], axis=1)
```

```python
#cria retornos do ativo e NAV começando em 100
df['Retornos'] = df.Close.pct_change()
#nav

df['NAV'] = (1+ df.Retornos).cumprod()*100 
df.NAV.iloc[0]=100

#df.head(50)
```

```python
df.NAV.plot()
```

```python
#exporta usando método .to_excel
df.to_excel('Exportação ^BVSP.xlsx')
#retornos.to_excel('Retornos BVSP.xlsx')
```
