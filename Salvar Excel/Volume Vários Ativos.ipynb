```python
import pandas as pd
import pandas_datareader as reader
import datetime as dt
```

```python
#cria lista com codigos dos ativos a serem baixados

tickers = ['vale3.sa','goll4.sa','brkm5.sa','mrfg3.sa','wege3.sa']
start = '01-01-2015'
end = dt.datetime.today()
```

```python
#usa pandas_datareader para baixar apenas volume dos ativos da lista tickers

df = reader.get_data_yahoo(tickers, start, end)['Volume']

df
```
